```python
one_scores = {}

one_scores['extroversion_score'] =  one[0] - one[1] + one[2] - one[3] + one[4] - one[5] + one[6] - one[7] + one[8] - one[9]
one_scores['nervousness_score'] =  one[0] - one[1] + one[2] - one[3] + one[4] + one[5] + one[6] + one[7] + one[8] + one[9]
one_scores['compatibility_score'] =  -one[0] + one[1] - one[2] + one[3] - one[4] - one[5] + one[6] - one[7] + one[8] + one[9]
one_scores['meticulousness_score'] =  one[0] - one[1] + one[2] - one[3] + one[4] - one[5] + one[6] - one[7] + one[8] + one[9]
one_scores['creativity_score'] =  one[0] - one[1] + one[2] - one[3] + one[4] - one[5] + one[6] + one[7] + one[8] + one[9]
```

```python
one_scores
```

```python
# ********************
```

```python
all_types = {'one':one, 'two': two, 'three' :three, 'four':four, 'five':five, 'six': six, 'seven': seven, 'eight': eight,
             'nine': nine, 'ten': ten}

all_types_scores ={}

for name, personality_type in all_types.items():
    personality_trait = {}

    personality_trait['extroversion_score'] =  personality_type[0] - personality_type[1] +personality_type[2] - personality_type[3] + personality_type[4] - personality_type[5] +personality_type[6] - personality_type[7] + personality_type[8] -personality_type[9]
    personality_trait['nervousness_score'] =  personality_type[0] - personality_type[1] + personality_type[2] -personality_type[3] + personality_type[4] + personality_type[5] + personality_type[6] + personality_type[7] + personality_type[8] + personality_type[9]
    personality_trait['compatibility_score'] =  -personality_type[0] +personality_type[1] - personality_type[2] + personality_type[3] - personality_type[4] - personality_type[5] + personality_type[6] - personality_type[7] + personality_type[8] + personality_type[9]
    personality_trait['meticulousness_score'] = personality_type[0] - personality_type[1] + personality_type[2] -personality_type[3] +personality_type[4] - personality_type[5] +personality_type[6] -personality_type[7] + personality_type[8] + personality_type[9]
    personality_trait['creativity_score'] =  personality_type[0] -personality_type[1] + personality_type[2] - personality_type[3] + personality_type[4] - personality_type[5] +personality_type[6] + personality_type[7] + personality_type[8] + personality_type[9]
    
    all_types_scores[name] = personality_trait
```

```python
all_types_scores
```

```python
# ************************
```

```python
all_extroversion = []
all_nervousness =[]
all_compatibility =[]
all_meticulousness =[]
all_creativity =[]

for personality_type, personality_trait in all_types_scores.items():
    all_extroversion.append(personality_trait['extroversion_score'])
    all_nervousness.append(personality_trait['nervousness_score'])
    all_compatibility.append(personality_trait['compatibility_score'])
    all_meticulousness.append(personality_trait['meticulousness_score'])
    all_creativity.append(personality_trait['creativity_score'])
    
    
```

```python
all_extroversion_normalized = (all_extroversion-min(all_extroversion))/(max(all_extroversion)-min(all_extroversion))
all_nervousness_normalized = (all_nervousness-min(all_nervousness))/(max(all_nervousness)-min(all_nervousness))
all_compatibility_normalized = (all_compatibility-min(all_compatibility))/(max(all_compatibility)-min(all_compatibility))
all_meticulousness_normalized = (all_meticulousness-min(all_meticulousness))/(max(all_meticulousness)-min(all_meticulousness))
all_creativity_normalized = (all_creativity-min(all_creativity))/(max(all_creativity)-min(all_creativity))
```

```python
all_extroversion_normalized
```

```python
all_nervousness_normalized
```

```python
counter = 0

normalized_all_types_scores ={}

for personality_type, personality_trait in all_types_scores.items():
    normalized_personality_trait ={}
    normalized_personality_trait['extroversion_score'] = all_extroversion_normalized[counter]
    normalized_personality_trait['nervousness_score'] = all_nervousness_normalized[counter]
    normalized_personality_trait['compatibility_score'] = all_compatibility_normalized[counter]
    normalized_personality_trait['meticulousness_score'] = all_meticulousness_normalized[counter]
    normalized_personality_trait['creativity_score'] = all_creativity_normalized[counter]
    
    normalized_all_types_scores[personality_type] = normalized_personality_trait
    
    counter+=1
```
